```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('final merged data.csv')
df
```

```python
agewalkmodel = sm.OLS(df['Walk Score'], df['Median Age']).fit()
predictions = agewalkmodel.predict(df['Median Age']) # make the predictions by the model

# Print out the statistics
agewalkmodel.summary()
```

```python
incomewalkmodel = sm.OLS(df['Walk Score'], df['Median Household Income']).fit()
predictions = incomewalkmodel.predict(df['Median Household Income']) # make the predictions by the model

# Print out the statistics
incomewalkmodel.summary()
```

```python
popwalkmodel = sm.OLS(df['Walk Score'], df['Population']).fit()
predictions = popwalkmodel.predict(df['Population']) # make the predictions by the model

# Print out the statistics
popwalkmodel.summary()
```
