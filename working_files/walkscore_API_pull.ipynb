```python
# Example test
url_string = 'http://api.walkscore.com/score?format=json&lat=%f&lon=%f&transit=1&bike=1&wsapikey=4e9a14b4cda5f4c809d23bf0f13235d4'

for index, row in df.iterrows():
    new_string = url_string % (row['lat'],row['lon'])
    #print(new_string)
    response = req.get(new_string)
    jsoner = response.json()
    try:
        df.set_value(index,'walk score',jsoner["walkscore"])
    except:
        print('no walk score data')
    try:
        df.set_value(index,'transit score',jsoner["transit"]['score'])
    except:
        print('no transit score data')
    try:    
        df_test.set_value(index,'bike score',jsoner["bike"]['score'])
    except: 
        print('no bike score data')
    
```
